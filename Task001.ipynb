```python
import pandas as pd

# df = pd.read_csv("data/POPS.csv",thousands=',', decimal=',')
# print("First 5 rows:", end='')
# print("First 5 rows:")

df = pd.read_csv("data/POPS.csv", index_col=0)
print("First 5 rows:")
df.head()
```

```python
df.dtypes
```

# Data Cleaning
* Remove commas
* Convert from str to int

```python
df['Population'] = df['Population'].str.replace(',', '')
df['Metropolitan'] = df['Metropolitan'].str.replace(',', '')
```

```python
df['Population'] = df['Population'].astype(int)
df['Metropolitan'] = df['Metropolitan'].astype(int)
```

```python
print("Descriptive Statistics:")
df.describe()
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

# Maximum value in data

```python
print("Maximum Population: ", df['Population'].max())
print("Maximum Sex Ration: ", df['Sexratio'].max())
print("Maximum Literacy: ", df['Literacy'].max())
print("Maximum Metropolitan: ", df['Metropolitan'].max())
```

# Minimum value in data

```python
print("Minimum Population: ", df['Population'].min())
print("Minimum Sex Ration: ", df['Sexratio'].min())
print("Minimum Literacy: ", df['Literacy'].min())
print("Minimum Metropolitan: ", df['Metropolitan'].min())
```

# Average value in data

```python
print("Average Population: ", df['Population'].mean())
print("Average Sex Ration: ", df['Sexratio'].mean())
print("Average Literacy: ", df['Literacy'].mean())
print("Average Metropolitan: ", df['Metropolitan'].mean())
```

# State wise population

```python
state_wise_population = df.groupby('State')['Population'].sum().sort_values(ascending=False)
state_wise_population
```

# State wise avg literacy rate

```python
state_wise_avg_lit = df.groupby('State')['Literacy'].mean().sort_values(ascending=False)
state_wise_avg_lit
```
